# Research Question

How does a country's supply of animal and plant-based products affect their life expectancy?



## Background and Prior Work

Several factors influence an individual's lifespan, including genetics, environmental conditions, behavior, and diet. Genetics determines our susceptibility to certain diseases and influences our body’s ability to repair itself (Smulders and Deelen, 2023)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Environmental factors like air quality, access to healthcare, and socioeconomic status also play a role in longevity (Passarino et al., 2016)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Behaviors such as physical activity, smoking, alcohol consumption, and stress management significantly impact health outcomes (Wang et al., 2023)<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Diet, however, is one of the most modifiable factors affecting lifespan. A balanced diet can prevent chronic diseases, reduce inflammation, and support overall health (Caprara, 2018<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1); Willcox et al., 2007<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)). The goal of our project is to investigate how a country’s production of different foods (plant-based or animal-based) affects life expectancy-specific. We aim to identify which of these diets may help reduce the risk of age-related diseases and promote longevity.

A study by Caprara (2018) <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) analyzed how different diets reduce non-communicable diseases (NCDs). NCDs are diseases that do not transfer from person to person through infectious agents (heart disease, stroke, cancer, diabetes, and chronic lung disease). In the study, Caprara examined seven different dietary patterns to assess their potential in reducing the prevalence and impact of these conditions. The diets studied included the Mediterranean diet, known for its emphasis on healthy fats, fruits, and vegetables; the Japanese diet, which is rich in fish, rice, and plant-based foods; the vegetarian diet, which eliminates animal products; the new Nordic diet, which focuses on locally sourced, seasonal foods; the fast-mimicking diet, which simulates the effects of fasting while still providing essential nutrients; and the caloric restriction diet, which involves reducing calorie intake without malnutrition.These diets, common in Blue Zones where people live longer, share key features: high intake of plant-based foods, whole grains, unsaturated fats, and moderate protein, with a focus on omega-3s and antioxidants. They also limit red meat, processed foods, and sugars. Caprara also claimed that adopting a plant-based diet, alongside intermittent fasting and active living, can significantly reduce NCD risk and promote a longer, healthier life.

A study by Willcox et al. (2007) <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) explored the link between caloric restriction (CR) and the logn life span of the Okinawan population, specifically those aged 65 and older. The research found that Okinawans have some of the longest life expectancies in the world, and their longevity can be attributed to mild, long-term caloric restriction. This population historically consumed a nutrient-dense, low-calorie diet, combined with high physical activity levels, which may have contributed to lower mortality from age-related diseases and extended survival. The study suggests that energy deficits in early life led to adaptations that promoted healthier aging and a longer lifespan.

## Life Expectancy

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 

#read in dataset and skip 4 rows of metadata at the top
life_exp = pd.read_csv('life_expectancy.csv', skiprows=4)
life_exp

#drop unnecessary columns and columns with no data
life_exp = life_exp.drop(columns=['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
life_exp

#convert individuals years to observations and not columns
life_exp = life_exp.melt(id_vars=["Country Name", "Country Code"], 
                  var_name="Year", 
                  value_name="Life Expectancy")

#convert 'Year' to integer
life_exp['Year'] = life_exp['Year'].astype(int)
life_exp.dtypes

#filter to observations after 2010 to match our other dataset
life_exp = life_exp[life_exp['Year'] >= 2010]
life_exp

#reset index back to 0
life_exp.reset_index(drop=True, inplace=True)
life_exp
```

## Death Rate

```python
#read in dataset and skip 4 rows of metadata at the top
death_rate = pd.read_csv('death_rate.csv', skiprows=4)
death_rate

#drop unnecessary columns and columns with no data
death_rate = death_rate.drop(columns=['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
death_rate

#convert individuals years to observations and not columns
death_rate = death_rate.melt(id_vars=["Country Name", "Country Code"], 
                  var_name="Year", 
                  value_name="Death Rate")

#convert 'Year' to integer
death_rate['Year'] = death_rate['Year'].astype(int)
death_rate.dtypes

#filter to observations after 2010 to match our other dataset
death_rate = death_rate[death_rate['Year'] >= 2010]
death_rate

#reset index back to 0
death_rate.reset_index(drop=True, inplace=True)
death_rate

#merge this with our life expectancy dataset
health = pd.merge(life_exp, death_rate[['Country Code', 'Year', 'Death Rate', 'Country Name']], 
                     on=['Country Name', 'Country Code', 'Year'], how='inner')
health
```
