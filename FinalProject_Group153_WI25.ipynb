# Abstract

This study investigates the relationship between a country's supply of animal versus plant-based  food products and its life expectancy. Using data from the Food and Agriculture Organization (FAO) and the World Bank Group, our group analyzed the per capita availability of food and life expectancy across 3458 countries from 2010 to 2022. The goal of this project was to assess whether countries with higher supplies of plant-based foods, such as fruits and vegetables, had longer life expectancies compared to those with greater availability of animal-based foods. We hypothesized that plant-based food supply would correlate more strongly with higher life expectancy than animal-based food supply. This was due to our knowledge of Blue Zones, which are areas around the world that consistently show high life expectancies, attributed to their residents' diets that are rich in plant-based foods, healthy fats, and antioxidants. We initially believed that the nutrients in these diets, such as omega-3 fatty acids and antioxidants, contribute to reduced inflammation, improved cardiovascular health, and a lower incidence of chronic diseases, all of which may lead to longer lifespans.

However, our results revealed some unexpected findings. We observed that animal fat had the strongest positive correlation with life expectancy—each additional gram per day was associated with about a 0.1745-year increase—plant-based foods showed more mixed results. Plant calories were linked to a slight increase in life expectancy, while plant fat and plant protein had small negative effects. These findings diverged from our hypothesis, suggesting that while animal fat consumption may play a significant role in longevity, other factors such as lifestyle choices, genetics, and environmental influences likely contribute to a more complex relationship between diet and life expectancy. When utilizing a regression model, we found about a 65% of variation in our data which is strong but that still leaves room for lots of uncertainty.

## Background and Prior Work


Several factors influence an individual's lifespan, including genetics, environmental conditions, behavior, and diet. Genetics determines our susceptibility to certain diseases and influences our body’s ability to repair itself (Smulders and Deelen, 2023)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Environmental factors like air quality, access to healthcare, and socioeconomic status also play a role in longevity (Passarino et al., 2016)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Behaviors such as physical activity, smoking, alcohol consumption, and stress management significantly impact health outcomes (Wang et al., 2023)<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Diet, however, is one of the most modifiable factors affecting lifespan. A balanced diet can prevent chronic diseases, reduce inflammation, and support overall health (Caprara, 2018<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1); Willcox et al., 2007<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)). The goal of our project is to investigate how a country’s production of different foods (plant-based or animal-based) affects life expectancy-specific. We aim to identify which of these diets may help reduce the risk of age-related diseases and promote longevity.

A study by Caprara (2018) <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) analyzed how different diets reduce non-communicable diseases (NCDs). NCDs are diseases that do not transfer from person to person through infectious agents (heart disease, stroke, cancer, diabetes, and chronic lung disease). In the study, Caprara examined seven different dietary patterns to assess their potential in reducing the prevalence and impact of these conditions. The diets studied included the Mediterranean diet, known for its emphasis on healthy fats, fruits, and vegetables; the Japanese diet, which is rich in fish, rice, and plant-based foods; the vegetarian diet, which eliminates animal products; the new Nordic diet, which focuses on locally sourced, seasonal foods; the fast-mimicking diet, which simulates the effects of fasting while still providing essential nutrients; and the caloric restriction diet, which involves reducing calorie intake without malnutrition.These diets, common in Blue Zones where people live longer, share key features: high intake of plant-based foods, whole grains, unsaturated fats, and moderate protein, with a focus on omega-3s and antioxidants. They also limit red meat, processed foods, and sugars. Caprara also claimed that adopting a plant-based diet, alongside intermittent fasting and active living, can significantly reduce NCD risk and promote a longer, healthier life.

A study by Willcox et al. (2007) <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) explored the link between caloric restriction (CR) and the logn life span of the Okinawan population, specifically those aged 65 and older. The research found that Okinawans have some of the longest life expectancies in the world, and their longevity can be attributed to mild, long-term caloric restriction. This population historically consumed a nutrient-dense, low-calorie diet, combined with high physical activity levels, which may have contributed to lower mortality from age-related diseases and extended survival. The study suggests that energy deficits in early life led to adaptations that promoted healthier aging and a longer lifespan.

The FAOSTAT food supply dataset provides information on daily per capita availability of calories, protein, and fat across different countries and years and separated by plant or animal-based foods. The key variables include Food Supply (kcal/capita/day), which measures the average caloric intake available per person per day, Protein Supply (g/capita/day) and Fat Supply (g/capita/day), which quantify the availability of macronutrients. These metrics, reported as numerical values (float64), serve as proxies for national dietary patterns and potential nutritional adequacy. While they do not directly measure individual consumption, they reflect broader trends in the types of food available and certain macronutrient intakes within a country. The data is already tidy and does not contain any null values, so to clean it we just need to drop the columns that are not important for our analysis. 

The Life Expectancy and Death Rate datasets provided by the World Bank Group are measures of the average life expectancy at birth and the death rate per 1000 people. Each dataset has observations at the individual country level and per year. The measure of life expectancy is in years and death rate is the number of deaths per 1000 people in a country. These data needed lots of tidying and cleaning, first by dropping the metadata at the top that interfered with reading in the files. Then we had to drop unnecessary columns, convert data to a different type, melt the dataframes, filter to data from 2010-2022 to match the FAOSTAT data, and finally merge the two. We merged these two to create a comprehensive health dataset that had both life expectancy and death rate. 

## Food and Agriculture Supply

```python
#importing necessary packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#upload dataset with pandas
fao = pd.read_csv('FAOSTAT_data.csv')

#drop columns we won't use for our analysis
fao = fao.drop(columns=['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Item Code (FBS)', 'Year Code', 'Flag', 'Flag Description', 'Note'])

#rename the 'Area' column
fao.rename(columns={"Area": "Country Name"}, inplace=True)
fao
```

## Life Expectancy

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 

#read in dataset and skip 4 rows of metadata at the top
life_exp = pd.read_csv('life_expectancy.csv', skiprows=4)
life_exp

#drop unnecessary columns and columns with no data
life_exp = life_exp.drop(columns=['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
life_exp

#convert individuals years to observations and not columns
life_exp = life_exp.melt(id_vars=["Country Name", "Country Code"], 
                  var_name="Year", 
                  value_name="Life Expectancy")

#convert 'Year' to integer
life_exp['Year'] = life_exp['Year'].astype(int)
life_exp.dtypes

#filter to observations after 2010 to match our other dataset
life_exp = life_exp[life_exp['Year'] >= 2010]
life_exp

#reset index back to 0
life_exp.reset_index(drop=True, inplace=True)
life_exp
```

## Death Rate

```python
#read in dataset and skip 4 rows of metadata at the top
death_rate = pd.read_csv('death_rate.csv', skiprows=4)
death_rate

#drop unnecessary columns and columns with no data
death_rate = death_rate.drop(columns=['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
death_rate

#convert individuals years to observations and not columns
death_rate = death_rate.melt(id_vars=["Country Name", "Country Code"], 
                  var_name="Year", 
                  value_name="Death Rate")

#convert 'Year' to integer
death_rate['Year'] = death_rate['Year'].astype(int)
death_rate.dtypes

#filter to observations after 2010 to match our other dataset
death_rate = death_rate[death_rate['Year'] >= 2010]
death_rate

#reset index back to 0
death_rate.reset_index(drop=True, inplace=True)
death_rate

#merge this with our life expectancy dataset
health = pd.merge(life_exp, death_rate[['Country Code', 'Year', 'Death Rate', 'Country Name']], 
                     on=['Country Name', 'Country Code', 'Year'], how='inner')

#drop and rename columns
health = health.drop(columns=['Country Code'])
health.rename(columns={'Country Name' : 'Country'}, inplace=True)
health
```

# Results

## Exploratory Data Analysis



### Plant vs Animal Products on Life Expectancy

```python
#looking for any null values
display(fao[fao.isnull().any(axis=1)])

#looking at summary statistics of the dataframe
print(fao.info())
print(fao.describe())
print(fao['Element'].unique())
```

```python
#creating a pivot table to better organize our data
df_pivot = fao.pivot_table(index=["Country Name", "Year", "Item"], 
                           columns="Element", 
                           values="Value", 
                            aggfunc="sum").reset_index()

#renaming the columns
df_pivot.columns = ['Country', 'Year', 'Item', 'Fat', 'Calories', 'Protein']
df_pivot

#now pivot again so that we can separate animal vs plants for calories, fats, and protein
df_final = df_pivot.pivot_table(index=["Country", "Year"], 
                                columns="Item", 
                                values=["Calories", "Protein", "Fat"], 
                                aggfunc="sum").reset_index()

#rename columns again
df_final.columns = ["Country", "Year", 
                    "Animal_Calories", "Plant_Calories", 
                    "Animal_Protein", "Plant_Protein", 
                    "Animal_Fat", "Plant_Fat"]
df_final
```

```python
#renaming Health dataframe column to prep for merge
health.rename(columns={"Country Name": "Country"}, inplace=True)

#merging our FAO dataframe with the Health dataframe
df_merged = pd.merge(df_final, 
                     health[["Country", "Year", "Life Expectancy", "Death Rate"]], 
                     on=["Country", "Year"], how="inner")
```

```python
#creating a correlation matrix 
corr_matrix = df_merged.drop(columns=["Country", "Year"]).corr()

#creating a heatmap to visualize correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Between Types of Food and Life Expectancy")
plt.show()
```

Looking at the heatmap of our correlation matrix we can see a clear relationship between the food source and life expectancy/death rate. Interestingly enough, we see that animal calories, animal protein, and animal fat all have a strong positive correlation with life expectancy. This would mean that animal-based products are associated with longer lifespans. Meanwhile, plant calories, plant protein and plant fat all have relatively weak correlations with life expectancy. The correlations are contradictory to our hypothesis which makes us question if there are other confounding variables affecting the results? 

```python
# creating a scatterplot of animal-based products on life expectancy
sns.scatterplot(x=df_merged["Animal_Calories"], y=df_merged["Life Expectancy"])
plt.title("Animal-Based Calories vs. Life Expectancy")
plt.show()

# creating a scatterplot of plant-based products on life expectancy
sns.scatterplot(x=df_merged['Plant_Calories'], y=df_merged['Life Expectancy'])
plt.title("Plant-Based Calories vs. Life Expectancy")
plt.show()
```

These scatterplots also show a strong positive correlation between life expectancy and animal calories. There is also a positive correlation between plant calories and life expectancy. For animal calories, we can see that there are far more values in the low-life expectancy range as compared to the values for plant calories.

```python
# comparing plant protein vs animal protein on life expectancy
sns.scatterplot(x=df_merged["Animal_Protein"], y=df_merged["Life Expectancy"])
plt.title("Animal-Based Protein vs. Life Expectancy")
plt.show()

sns.scatterplot(x=df_merged["Plant_Protein"], y=df_merged["Life Expectancy"])
plt.title("Plant-Based Protein vs. Life Expectancy")
plt.show()
```

Here we seem to see a stronger, positive relationship between plant-protein and life expectancy as an increase in plant-protein leads to an increase in life expectancy. However, animal-based protein seems to have a stronger effect still.

```python
# now comparing plant fat vs animal fat on life expectancy
sns.scatterplot(x=df_merged["Animal_Fat"], y=df_merged["Life Expectancy"])
plt.title("Animal-Based Fat vs. Life Expectancy")
plt.show()

sns.scatterplot(x=df_merged["Plant_Fat"], y=df_merged["Life Expectancy"])
plt.title("Plant-Based Fat vs. Life Expectancy")
plt.show()
```

As before with calories, we see a stronger, more linear relationship between animal-based products and life expectancy than plant-based. This might suggest that although plant-based products still effect life expectancy, that animal-based products have a much stronger effect. 

Let's run a regression to better analyze how different types of food and the source affect life expectancy.

```python
X = df_merged[["Animal_Calories", "Plant_Calories", "Animal_Protein", "Plant_Protein", "Animal_Fat", "Plant_Fat"]]
X = sm.add_constant(X)  # Adds intercept
y = df_merged["Life Expectancy"]

# Run regression
model = sm.OLS(y, X).fit()
print(model.summary())
```

The regression tells us that the model predicts 65% of the variation in our data which is strong but that still leaves room for lots of uncertainty. Looking at the model, animal fat has the strongest correlation with life expectancy. A 1 gram increase per day of animal fat is associated with a 0.1745 year increase in life expectancy. Plant calories has a small prositive relationship, which tells us that a 1kcal increase is associated with a 0.0101 year life increase. Meanwhile, plant fat and plant protein both have small negative effects which means that an increase in the consumption of these foods is associated with a small decrease in life expectancy. Animal calories and animal protein are both not statistically significant, which means that when controlling for other factors they don't strongly predict life expectancy. These results are interesting, but still suggest that there are other factors impacting our data. Let's look at what countries concumer the most of these different foods and see if that data can tell us anything interesting. 

```python
#looking at which countries consume the most of each product
#filter the data for animal and plant products
animal_data = fao[fao['Item'] == 'Animal Products']
plant_data = fao[fao['Item'] == 'Vegetal Products']

#calculate total values for each animal nutritional element
animal_calories = animal_data[animal_data['Element'] == 'Food supply (kcal/capita/day)']
animal_protein = animal_data[animal_data['Element'] == 'Protein supply quantity (g/capita/day)']
animal_fat = animal_data[animal_data['Element'] == 'Fat supply quantity (g/capita/day)']

#calculate total values for each plant nutritional element
plant_calories = plant_data[plant_data['Element'] == 'Food supply (kcal/capita/day)']
plant_protein = plant_data[plant_data['Element'] == 'Protein supply quantity (g/capita/day)']
plant_fat = plant_data[plant_data['Element'] == 'Fat supply quantity (g/capita/day)']

#groupby country and sum up the values
animal_calories = animal_calories.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)
animal_protein = animal_protein.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)
animal_fat = animal_fat.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)

plant_calories = plant_calories.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)
plant_protein = plant_protein.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)
plant_fat = plant_fat.groupby('Country Name')['Value'].sum().reset_index().sort_values(by='Value', ascending=False)
```

# Discusison and Conclusion
Our study aimed to determine whether an animal-based diet (meat, dairy, and eggs) or a plant-based diet (fruits, vegetables, and grains) is more strongly associated with increased human lifespan. Contrary to our hypothesis, our findings indicate that an animal-based diet has a stronger correlation with longevity than a plant-based diet.

To reach this conclusion, we conducted exploratory data analysis using three datasets covering food and agricultural supply, life expectancy, and death rates across various countries. A heatmap analysis revealed a strong correlation between animal-based calories, protein, and fat with life expectancy, whereas plant-based food components showed much weaker associations. We also generated six scatter plots comparing life expectancy to different dietary components: animal-based calories, protein, and fat, as well as plant-based calories, protein, and fat. While all six showed a positive correlation, the association was notably stronger for animal-based foods, reinforcing the idea that these products may have a greater impact on longevity. To further quantify these relationships, we performed a regression analysis, which revealed that animal fat had the strongest positive correlation with life expectancy. Specifically, an increase of 1 gram per day of animal fat was associated with an increase of approximately 0.1745 years in life expectancy. However, the model did not provide a clear answer regarding the effects of other dietary variables, highlighting the complexity of dietary influences on longevity. Although our regression model explained 65% of the variation in life expectancy, a significant portion remains unexplained. Comparing the average life expectancy in countries that consumed more plant-based products vs those that consumed more animal-based products, we found that the countries with higher consumption of animal products had longer life expectancy. These results were contradictory to our hypothesis that a higher consumption of plant products would be associated with longer life expectancy while a higher consumption of animal products would be associated with a shorter life expectancy. This was particularly interesting to us because our hypothesis was backed by extensive research in the field of nutrition, which has consistently linked plant-based diets to longevity. This contradiction suggests that within our research, diet alone may not be a strong enough predictor of life expectancy.  

We must take into account the limitations of our study. There are various factors that can significantly impact life expectancy in certain countries other than diet, such as exercise, sanitation practices, smoking, alcohol, GDP, and many other factors. Since we were only measuring diet, including these factors in future research could very likely lead to different conclusions. Also due to the limitations in the data set we could only compare the calories, protein, and fat from plants and animals, but there are many other nutritional factors that could affect our research. 

The findings in this study could lead society to consider focusing their diet more on animal-based foods if in fact this can contribute to an increased lifespan. With the growing popularity and advertisement of plant-based foods, many may be misled into believing that plant-based diets are inherently "healthier" than those based on animal proteins. Taking this into consideration, it may lead to an impact on social awareness of misinformation being spread in our everyday lives on diet choices.
